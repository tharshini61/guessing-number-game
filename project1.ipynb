## level 1- guessing correct number

```python
user_input= int(input("enter a number:"))
guessed_number=int(input("enter a number 1 to 100:"))
if guessed_number== user_input:
    print("your guess is correct")
else:
    print("try again")
```

## level 2-guessing number checks it is higher or lesser 

```python
secret_number= int(input("enter a number:"))
guessed_number=int(input("enter a number 1 to 100:"))
if guessed_number== secret_number:
    print("your guess is correct")
elif guessed_number >=0:
  print("your guessed number is higher")
elif guessed_number<=100:
    print("your guessed number is lesser")
else:
    print("try again")
```

## level 3- random secret number

```python
import random 
secret_number=random.randint(1,100)
guessed_number=int(input("enter a number 1 to 100:"))
if guessed_number== secret_number:
    print("your guess is correct")
elif guessed_number >=0:
  print("your guessed number is higher")
elif guessed_number<=100:
    print("your guessed number is lesser")
else:
    print("try again")

```

## level 4- should be guessed in 5 attempts

```python
# import random  
secret_number=random.randint(1,100)
print("you have 5 chances to guess the secret number.")
for i in range(1,6):
    guessed_number=int(input(f"try {i} in attempts - enter a number 1 to 100:"))
    if guessed_number == secret_number:
        print("your guess is correct")
        break
    elif guessed_number > secret_number:
        print("your guessed number is higher")
    else:
        print("your guessed number is lesser")
        if i==5 and guessed_number!= secret_number:
            print(f"game is over try again!! my secret number is{ secret_number}")



```
